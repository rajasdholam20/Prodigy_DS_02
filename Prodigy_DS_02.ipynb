```python
import pandas as pd
import matplotlib.pyplot as np
#importing DataSet
```

Data cleaning

```python
df = pd.read_csv("titanic.csv")
df
#Reading & displaying The DataSet
```

```python
df.dropna
df
```

```python
df.isnull().sum()
# checking for the null values & observing there are 177 null values in Age & 687 null values in Cabin
```

```python
# now filling the null values of age with median of age
df['Age'].fillna(df['Age'].median(),inplace = True)
df
```

```python
df['Age'].isnull().sum()
#now we can observe that the data column Age has Zero null values.
```

```python
#now filling the null values of Column Cabin with the mode of Cabin
df['Cabin'].fillna(df['Cabin'].mode()[0], inplace=True)
df
```

```python
df['Cabin'].isnull().sum()
#after imuting the values we can see that the null values in Column are 0 replaced by the mode value
```

```python
df['Embarked'].isnull().sum()
#there are 2 missing values in th Column Embarked which will be filled now 
```

```python
df['Embarked'].fillna(df['Embarked'].mode()[0], inplace = True)
df
#Replacing the missing values with the mode of the values
```

```python
#Now checking the null valuyes if replaced succesfully
df['Embarked'].isnull().sum()
```

```python
#The missing values are replaced succesfully
```

```python
#Now Checking for the Duplicate Values

df.duplicated().sum()
```

```python
#Now Drping al the duplicate values

df.drop_duplicates()
```

```python
#checking for duplications if succesfull
df.duplicated().sum
```

Co-realation between different Variables

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

#excluding non numerical data or Including only numerical data to show the correlation
n_df = df.select_dtypes(include=['number'])

corr_matrix = n_df.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, fmt='.2f', cmap='coolwarm', cbar=True)
plt.title('Correlation Matrix')
plt.show()
```

```python
# Distribution of categorical data
fig, axes = plt.subplots(2, 2, figsize=(14, 12))
plt.subplots_adjust(hspace=0.5)

sns.countplot(x='Survived', data=df, ax=axes[0, 0])
axes[0, 0].set_title('Survival Count')

sns.countplot(x='Pclass', data=df, ax=axes[0, 1])
axes[0, 1].set_title('Passenger Class Distribution')

sns.countplot(x='Sex', data=df, ax=axes[1, 0])
axes[1, 0].set_title('Gender Distribution')

sns.countplot(x='Embarked', data=df, ax=axes[1, 1])
axes[1, 1].set_title('Embarkation Point Distribution')

'''sns.countplot(x='Age' , data=df, ax=axes[2,0])
axes[2,0].set_title('Age Distribution')

sns.countplot(x='Fare', data=df, ax=axes[2,1])
axes[2,1].set_title('Fare Distribut')'''
plt.show()
```
